## Handling Outliers


```python

q1 = data['temperature'].quantile(0.25)
q2 = data['temperature'].quantile(0.75)
IQR = q2 - q1
no_outliers = data[(data['temperature'] > q1 - 1.5 * IQR) & (data['temperature'] < q2 + 1.5 * IQR)]
```

```python
no_outliers
```
