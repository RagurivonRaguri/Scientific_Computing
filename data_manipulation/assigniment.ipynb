```python
import pandas as pd

data=pd.read_csv("./IMDb 2024 Movies TV Shows.csv")
data
```

```python
data.isnull().sum()
```

```python
import re # This is the regex Python library used for searching and manipulating strings

numerical_columns = ["Budget", "Revenue", "Run_Time", "Vote_Average", "Vote_Count"]
# Clean the Budget column using regex
data[numerical_columns]=data[numerical_columns].astype(str) 
data[numerical_columns]=data[numerical_columns].map(lambda x: re.sub(r'[^0-9.]', '', x)) # Remove non-numeric characters except dot

# Convert to numeric
data[numerical_columns] = data[numerical_columns].apply(pd.to_numeric, errors='coerce') 
data
```

```python
data.isnull().sum()
```

```python
# Fill missing values with the mean
data['Budget'] = data['Budget'].fillna(data['Budget'].mean()) 
data['Run_Time'] = data['Run_Time'].fillna(data['Run_Time'].mean())
data
```

```python
data['Budget'].mean()
```

```python
data['Run_Time'].mean()
```

Statistical Data:
-First, Second and Third Quartile
-Interquartile Range

```python
quartile1 = data['Budget'].quantile(0.25)
quartile2 = data['Budget'].quantile(0.5)
quartile3 = data['Budget'].quantile(0.75)
IQR = quartile3 - quartile1
quartile1, quartile2, quartile3
IQR
```

Outliers

```python
outliers = data[(data['Budget'] > quartile1 - 5 * IQR) & (data['Budget'] < quartile3 + 5 * IQR)]
outliers
```

Data Representation

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Scatter plot

```python
sns.scatterplot(x="Budget", y="Revenue", size="Vote_Average", data=data)
plt.show()
```

Histogram

Histogram of Revenue against Budgets

```python
data.plot(kind='hist', x='Budget', y='Revenue', bins=5, color='green')
plt.title('Histogram of Revenue against Budgets')
plt.ylabel('Revenue')
plt.xlabel('Budget')
plt.show()
```

Correlation Heatmap

```python

data_numeric = data[numerical_columns].copy()

# Compute correlation matrix
corr = data_numeric.corr()

# create heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap (Numerical Data)")
plt.show()
```
